```python
import mysql.connector

# Connect to the database
conn = mysql.connector.connect(
    host="localhost",       # Database host
    user="root",            # Database username
    password="", # Database password
    database="city-traffic-and-accident-pattern-analysis"  # Database name
)

cursor = conn.cursor()

# Run a query
cursor.execute("SELECT weather_condition, lighting_condition, COUNT(*) as total_accidents FROM `traffic_accidents` GROUP BY weather_condition, lighting_condition ORDER by total_accidents ASC;")

# Fetch results
row = cursor.fetchall()
dfAnn = pd.DataFrame(row)

# Close connection
cursor.close()
conn.close()

dfAnn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    df[column] = le.fit_transform(df[column])
```

```python
df
```

```python
y = df[['crash_type']]
```

```python
y
```

```python
df1 = df.drop(columns=['crash_type'], axis = 1)
df1
```

```python
for column in df1.columns:
    if column.startswith('injuries') or column.endswith('injury'):
        df1 = df1.drop(columns=[column], axis = 1)
df1
```

```python
x = df1
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
y_pred
```

```python
pd.DataFrame(y_pred)
```

```python
lr.score(x_test, y_test)
```

```python
from sklearn.metrics import accuracy_score, r2_score, precision_score, recall_score, f1_score
accuracy_score(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
```

```python
print(report)
```
